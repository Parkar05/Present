# Hello. Here Are Some Insights About the Developers Data

### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the Data

```python
df = pd.read_csv("ANALYSIS.csv")
df.head(3)
```

## Lets Begin With Insights 

### Distribution of People's current job positions on Linkedln

```python
# Distribution of current job positions
plt.figure(figsize=(10, 6))  
sns.countplot(x='Current Position', data=df)
plt.title('Distribution of Current Job Positions')
plt.xticks(rotation=45, ha='right')  
plt.xlabel('Current Position')
plt.ylabel('Count')
plt.tight_layout()  
plt.show()
```

### Position assigned by us

```python
# Distribution of positions
plt.figure(figsize=(10, 6))  
sns.countplot(x='Position ', data=df)
plt.title('Distribution of Positions')
plt.xticks(rotation=45, ha='right')  
plt.xlabel('Position ')
plt.ylabel('Count')
plt.tight_layout() 
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Position ', data=df, palette='viridis', order=df['Position '].value_counts().index)
plt.title('Distribution of Positions', fontsize=16, fontweight='bold')
plt.xlabel('Position ', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Showing Experience of people we have Using scatter plot and bar plot

```python
# Scatter plot of experience vs. position
plt.figure(figsize=(16, 6))
sns.scatterplot(x='Experience ', y='Position ', data=df)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title('Experience vs. Position')
plt.xlabel('Experience (years)')
plt.ylabel('Position')
plt.show()
```

### Barplot

```python
# Convert 'Experience' column to numeric
df['Experience '] = df['Experience '].str.extract(r'(\d+)').astype(float)

# Now, plot the barplot
plt.figure(figsize=(14, 6))
sns.barplot(x='Position ', y='Experience ', data=df, ci=None, estimator=np.mean, palette='viridis')
plt.title('Average Experience by Position', fontsize=16, fontweight='bold')
plt.xlabel('Position ', fontsize=14)
plt.ylabel('Average Experience', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Showing Number of Developers

```python
# Set the style of the plot
sns.set_style("whitegrid")

# Create a bar plot of the counts of each position
plt.figure(figsize=(10, 6))
sns.countplot(x='Position ', data=df, palette='viridis')
plt.title('Count of Each Position', fontsize=16)
plt.xlabel('Position ', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Some insights related to Experience of employee, position and name 

```python
# Distribution of Experience
plt.figure(figsize=(10, 6))
df['Experience '].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Experience')
plt.xlabel('Experience (years)')
plt.ylabel('Frequency')
plt.show()

# Average Experience by Position
plt.figure(figsize=(12, 6))
df.groupby('Position ')['Experience '].mean().sort_values().plot(kind='barh', color='lightgreen')
plt.title('Average Experience by Position')
plt.xlabel('Average Experience (years)')
plt.ylabel('Position')
plt.show()

# Top 10 Most Experienced Employees
top_10 = df.nlargest(10, 'Experience ')[['Name ', 'Experience ']]
print("Top 10 Most Experienced Employees:")
print(top_10)

# Experience vs. Position (Box Plot)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Position ', y='Experience ', data=df, palette='coolwarm')
plt.title('Experience Distribution by Position')
plt.xlabel('Position ')
plt.ylabel('Experience (years)')
plt.xticks(rotation=45)
plt.show()
```

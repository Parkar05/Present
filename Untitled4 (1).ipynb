# Hello. Here Are Some Insights About the Developers Data

### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the Data

```python
df = pd.read_csv("ANALYSIS.csv")
df.head(3)
```

## Lets Begin With Insights 

### Distribution of current job positions on Linkedln

```python
# Distribution of current job positions
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.countplot(x='Current Position', data=df)
plt.title('Distribution of Current Job Positions')
plt.xticks(rotation=45, ha='right')  # Rotate labels and adjust alignment
plt.xlabel('Current Position')
plt.ylabel('Count')
plt.tight_layout()  # Adjust layout to prevent label cutoff
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Current Position', data=df, palette='viridis', order=df['Current Position'].value_counts().index)
plt.title('Distribution of Current Job Positions', fontsize=16, fontweight='bold')
plt.xlabel('Current Position', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
